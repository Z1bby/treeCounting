```python
from deepforest import main
from deepforest import get_data
import os
import matplotlib.pyplot as plt
```

```python
model = main.deepforest()
model.use_release()
```

```python
for i in range(0,13):
    img = model.predict_image(path="./Drone2k_" + str(i) + ".png", return_plot=True)
    plt.figure()
    plt.imshow(img[:,:,::-1])
```

```python
tree_count = 0
for i in range(0,13):
    tree_labels = model.predict_image(path="./Drone2k_" + str(i) + ".png")
    tree_count_segment = len(tree_labels)
    print(tree_count_segment)
    tree_count += tree_count_segment
print("Counted " + str(tree_count) + " trees")
```
